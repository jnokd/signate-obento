```python
#https://qiita.com/shnchr/items/bdd4810b3477aa977413
import warnings
warnings.filterwarnings("ignore")
from pycaret.classification import *
import pandas as pd
```

データを読み込む

```python
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
print("Train:",train.shape, "Test:",test.shape) #Size
```

```python
import pandas_profiling
train.profile_report()
```

前処理

```python
from pycaret.regression import *
exp1 = setup(train, target = 'y') 
```

```python
モデルを比較
```

```python
compare_models()
```

パラメータチューニング

```python
tuned_model = tune_model(estimator='et')
```

結果を可視化

```python
evaluate_model(tuned_model)
```

テストデータで予測

```python
final_lr = finalize_model(tuned_model)
pred = predict_model(final_lr, data=test)
pred.head()
```

SIGNATE提出用データ作成

```python
sample=pd.read_csv("sample.csv",header=None)

sample[1]=pred['Label']
sample.to_csv("submit_obento_{}.csv".format('pycaret'),index=None,header=None)
print(sample.head())
```
